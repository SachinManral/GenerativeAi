## Streaming and Batch

### Streaming
Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experiences, particularly for longer responses. Calling stream() returns an iterator that yields output chunks as they are produced. You can use a loop to process each chunk in real-time.

```python
model.invoke("Write me a 500 words paragraph on AI")
```

## Out will be in a flow

```python
for chunk in model.stream("Write me a 500 words paragraph on AI"):
    print(chunk.text, end="", flush=True)
```
